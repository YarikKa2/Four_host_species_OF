```python
from PIL import Image
import os
from pathlib import Path
```

```python
photo_path = os.path.dirname(os.getcwd()).replace('\\', '/') + '/Photoes/'
```

```python
# Cycle that turns .tif or .png files to .jpg for morphometry program
for i in os.listdir(photo_path): # go to root path
    
    #in root path open H&E folder and go through folders in it
    for j in range(len(os.listdir(photo_path+i+'/H&E'))): 
        print('--'*18+'\n'+f'Directory: {i}/H&E/{j+1}'+'\n'+'--'*18)
        
        
        # If file .png or .tif convert it into .jpg with PIL 
        for file in os.listdir(photo_path+i+'/H&E'+f'/{j+1}'):
            if file.lower().endswith('.png') or file.lower().endswith('.tif'):
            
                orig_file_path = photo_path+i+'/H&E'+f'/{j+1}/' + f'{file}' #Location of the original file 
                new_file_path = orig_file_path.split('.')[0] +'.jpg' #Future location of the tranformed file
                
                img = Image.open(orig_file_path).convert('RGB') #Convert to RGB and save as .jpeg
                img.save(new_file_path, 'JPEG', qulity=95)
                
                os.remove(orig_file_path) #Remove original file
                
    
        print(*os.listdir(photo_path+i+'/H&E'+f'/{j+1}'), sep='\n')
        print('==' *18+'\n')
```
